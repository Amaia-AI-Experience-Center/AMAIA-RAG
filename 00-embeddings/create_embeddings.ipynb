# Creación de Embeddings de Texto

Configuración del cliente OpenAI y carga de variables de entorno para generar embeddings.

```python
import os
from openai import AzureOpenAI
import dotenv

dotenv.load_dotenv()

# Initialize Azure OpenAI client
openai_client = AzureOpenAI(
    api_key=os.environ["AZURE_OPENAI_API_KEY"],
    api_version=os.environ["AZURE_OPENAI_API_VERSION"],
    azure_endpoint=os.environ["AZURE_OPENAI_ENDPOINT"]
)

EMBEDDING_DEPLOYMENT = os.environ["AZURE_OPENAI_DEPLOYMENT_EMBEDDING"]
print(f"Usando Azure OpenAI con modelo: {EMBEDDING_DEPLOYMENT}")
```

Generación de un embedding para el texto "hello world" y visualización de su dimensión y datos.

```python
embeddings_response = openai_client.embeddings.create(
    model=EMBEDDING_DEPLOYMENT,
    input="hello world",
)
embedding = embeddings_response.data[0].embedding

print(f"Dimensiones del embedding: {len(embedding)}")
print(f"Primeros 10 valores: {embedding[:10]}")
```
